```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 회사별 노트북 가격비교 
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

plt.barh(pc1['company'],pc1['price_euros'],0.8)



plt.xlabel("Price")
plt.ylabel("Company")
plt.show()
```
